# Reconhecimento Facial com OpenCV

Aplicamos um classificador utilizando a biblioteca OpenCV. <br>
Utilizando o arquivo haarcascade_frontalface_alt.xml fornecido pela propria OpenCV com os parametros de classificac√£o.

```python
# Importa o OpenCV
import cv2
```

```python
# Define o caminho para a imagem a ser analisada
imagePath = 'C:/tmp/foto.jpg'
```

```python
from IPython.display import Image
Image('C:/tmp/foto.jpg')
```

```python
# Define o caminho para o arquivo que possui regras de  regras de classificacao (haarcascade_frontalface_alt.xml)
arq_parametros = 'C:/tmp/haarcascade_frontalface_alt.xml'
```

```python
# Cria o classificador
faceCascade = cv2.CascadeClassifier(arq_parametros)
```

```python
# Le a imagem e convete para escala de cinza (grey)
image = cv2.imread(imagePath)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Detecta rostos na imagem
faces = faceCascade.detectMultiScale( # Fun√ß√£o generica que detecta objetos
    gray, # Imagem em escala de cinza... muitas opera√ß√µes no OpenCV s√£o feitas em escalada de cinza
    scaleFactor = 1.1, # ALgumas faces podem parecem maiores que outras, esse paremetro corrige isso
    minNeighbors = 5, # Define quantos objs s√£o detectados proximo ao objeto atual
    minSize = (30, 30))  # Tamanho de cada quadro
```

```python
print("Encontramos {0} rostos!!".format(len(faces)))
```

```python
# Desenha os quadros ao redor dos rostos
for (x, y, width, height) in faces:
    cv2.rectangle(image, (x, y), (x+width, y+height), (255, 0, 0), 10)
```

```python
# Gravando o resultado
cv2.imwrite('C:/tmp/resultado.jpg', image)
```

```python
print("Encontramos {0} rostos!!".format(len(faces_v2)))
```

```python
# Desenha os quadros ao redor dos rostos
for (x, y, width, height) in faces_v2:
    cv2.rectangle(image, (x, y), (x+width, y+height), (255, 0, 0), 10)
```
